## Project: Build a Traffic Sign Classifier

In this notebook, I implement a deep-learning classifier to classify German traffic signs. This project was part of the Udacity Self-Driving Car Nanodegree and was completed in June 2017. Udacity provided seriealized, labeled traffic sign data for students to work with.

## Loading The Data:

```python
# Load pickled data
import pickle

training_file = "data/train.p"
validation_file= "data/valid.p"
testing_file = "data/test.p"

with open(training_file, mode='rb') as f:
    train = pickle.load(f)
with open(validation_file, mode='rb') as f:
    valid = pickle.load(f)
with open(testing_file, mode='rb') as f:
    test = pickle.load(f)
    
X_train, y_train = train['features'], train['labels']
X_valid, y_valid = valid['features'], valid['labels']
X_test, y_test = test['features'], test['labels']
```

## Dataset Summary & Exploration

The pickled data is a dictionary with 4 key/value pairs:

- `'features'` is a 4D array containing raw pixel data of the traffic sign images, (num examples, width, height, channels).
- `'labels'` is a 1D array containing the label/class id of the traffic sign. The file `signnames.csv` contains id -> name mappings for each id.
- `'sizes'` is a list containing tuples, (width, height) representing the original width and height the image.
- `'coords'` is a list containing tuples, (x1, y1, x2, y2) representing coordinates of a bounding box around the sign in the image. **THESE COORDINATES ASSUME THE ORIGINAL IMAGE. THE PICKLED DATA CONTAINS RESIZED VERSIONS (32 by 32) OF THESE IMAGES**

## Preprocess the Data

```python
X_train, y_train = train['features'], train['labels']
X_valid, y_valid = valid['features'], valid['labels']
X_test, y_test = test['features'], test['labels']

datasets = [X_train, X_valid, X_test]
for (x, i) in zip(datasets, range(len(datasets))):
    # Convert to grayscale, treating all chanells as equally important
    x = np.mean(x, axis=3)

    # Normalize
    x = (x - np.mean(x)) / np.std(x)

    # Turn back into 4d tensor (so we can use tf.nn.conv2d)
    x = np.expand_dims(x, 3)

    # Convert from float64 to float32 (since that's what tensor flow wants)
    x = x.astype(np.float32)
    datasets[i] = x

X_train, X_valid, X_test = datasets
               
for x in [("training", X_train),("validation", X_valid),("test", X_test)]:
    print("The {:} dataset has mean {:0.2f} and standard deviation {:0.2f}".format(x[0], np.mean(x[1]), np.std(x[1])))

i = np.random.randint(0, n_train)
img = X_train[i, :, :, 0]
plt.imshow(img)
plt.suptitle("Example of a pre-processed image:", fontsize = 16)
plt.show()
```

## Build The Classifier

```python
import tensorflow as tf

def conv2d(x, W, b, strides=1):
    x = tf.nn.conv2d(x, W, strides=[1, strides, strides, 1], padding='VALID')
    x = tf.nn.bias_add(x, b)
    return x


def maxpool2d(x, k=2):
    return tf.nn.max_pool(
        x,
        ksize=[1, k, k, 1],
        strides=[1, k, k, 1],
        padding='VALID'
    )


def fullyconnected(x, W, b, dropout=0.5):
    x = tf.add(tf.matmul(x, W), b)
    x = tf.nn.dropout(x, keep_prob=dropout)
    return x


def roadsign_net(x, keep_prob):
    mu = 0
    sigma = 0.1

    # TWO PATHS BETWEEN THE INPUT AND THE OUTPUT, ONE USING 5x5 convolutions, one using 7x7

    # PATH 1: (using two 5x5 convolutions two 2 pools)

    # 1st convolution layer- Input = 32x32x1. Output = 28x28x6.
    wc11 = tf.Variable(tf.truncated_normal([5, 5, 1, 6], mu, sigma))
    bc11 = tf.Variable(tf.zeros([6]))
    conv11 = conv2d(x, wc11, bc11, 1)

    # relu Activation.
    conv11 = tf.nn.relu(conv11)

    # 1st pooling layer- Input = 28x28x6. Output = 14x14x6.
    pool11 = maxpool2d(conv11, 2)

    # 2nd convolutional layer- Input 14x14x16, Output = 10x10x16
    wc12 = tf.Variable(tf.truncated_normal([5, 5, 6, 16], mu, sigma))
    bc12 = tf.Variable(tf.zeros([16]))
    conv12 = conv2d(pool11, wc12, bc12, 1)

    # relu Activation.
    conv12 = tf.nn.relu(conv12)

    # 2nd pooling layer- Input = 10*10*16. Output = 5x5x16
    pool12 = maxpool2d(conv12, 2)

    # Flatten- Input = 5x5x16, Output = 400.
    pool12 = tf.contrib.layers.flatten(pool12)

    # Layer 3: Fully Connected. Input = 400. Output = 120.
    wf11 = tf.Variable(tf.truncated_normal([400, 120], mu, sigma))
    bf11 = tf.Variable(tf.zeros([120]))
    fc11 = fullyconnected(pool12, wf11, bf11, keep_prob)
    
    # Activation
    fc11 = tf.nn.relu(fc11)

    # Layer 4: Fully Connected. Input = 120. Output = 84.
    wf12 = tf.Variable(tf.truncated_normal([120, 84], mu, sigma))
    bf12 = tf.Variable(tf.zeros([84]))
    fc12 = fullyconnected(fc11, wf12, bf12, keep_prob)
    # Activation.
    
    fc12 = tf.nn.relu(fc12)

    # PATH 2: (using two 7x7 convolutions and one pool)

    # 1st convolution layer- Input = 32x32x1. Output = 26x26x6.
    wc21 = tf.Variable(tf.truncated_normal([7, 7, 1, 6], mu, sigma))
    bc21 = tf.Variable(tf.zeros([6]))
    conv21 = conv2d(x, wc21, bc21, 1)

    # relu Activation.
    conv21 = tf.nn.relu(conv21)

    # 1st pooling layer- Input = 26x26x6. Output = 13x13x6.
    pool21 = maxpool2d(conv21, 2)
    
    # 2nd convolutional layer- Input 13x13x6, Output = 7x7x16
    wc22 = tf.Variable(tf.truncated_normal([7, 7, 6, 16], mu, sigma))
    bc22 = tf.Variable(tf.zeros([16]))
    conv22 = conv2d(pool21, wc22, bc22, 1)

    # relu Activation.
    conv22 = tf.nn.relu(conv22)

    # Flatten- Input = 7x7x16, Output = 784.
    conv22 = tf.contrib.layers.flatten(conv22)

    # Layer 3: Fully Connected. Input = 784. Output = 120.
    wf21 = tf.Variable(tf.truncated_normal([784, 120], mu, sigma))
    bf21 = tf.Variable(tf.zeros([120]))

    fc21 = fullyconnected(conv22, wf21, bf21, keep_prob)
    # relu Activation
    fc21 = tf.nn.relu(fc21)

    # Layer 4: Fully Connected. Input = 120. Output = 84.
    wf22 = tf.Variable(tf.truncated_normal([120, 84], mu, sigma))
    bf22 = tf.Variable(tf.zeros([84]))

    fc22 = fullyconnected(fc21, wf22, bf22, keep_prob)
    # Activation.
    fc22 = tf.nn.relu(fc22)

    # Concatenate the two paths, input (84, 84), output 168
    combined_layer = tf.concat(axis=1, values=(fc12, fc22))

    # Now combine the two paths into a final fully connected layer:
    wfcomb = tf.Variable(tf.truncated_normal([168, 43], mu, sigma))
    bfcomb = tf.Variable(tf.truncated_normal([43]))

    logits = fullyconnected(combined_layer, wfcomb, bfcomb, keep_prob)
    return logits
```

## Explanation of Model Architecture:

The architecture of the network used to classify traffic signs is most easily understood as a modification of LeNet. (http://www.pyimagesearch.com/2016/08/01/lenet-convolutional-neural-network-in-python/)

Like in LeNet, the input passes through a 5x5 convolutional layer, followed by a maxpool with stride 2, followed by another 5x5 convolution, followed by a maxpool of stride 2, followed by two fully connected layers. (lines 30-71 in the model architecture cell)

Unlike LeNet there is a second path through the graph from the input to the output, which involves a 7x7 convolutional layer followed by maxpool with stride 2, followed by a 7x7 convolutional layer, followed by two fully connected layer. (lines 73-110 in the model architecture cell)

The two fully connected layers at the end of the graph are then concatenated together, and passed through one final fully connected layer which converts the signal into a 43 dimensional array (since there are 43 classes). (lines 112-121 in the model architecture cell).

All convolution and linear layers have a relu activation. Fully connected layers include dropout.

## Set Parameters

```python
x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
keep_prob = tf.placeholder(tf.float32)
one_hot_y = tf.one_hot(y, 43)

rate = 0.0005

EPOCHS = 20
BATCH_SIZE = 64
DROPOUT = 0.5  # Increase this to increase the keep probability
```

## Pipeline

```python
from sklearn.utils import shuffle

logits = roadsign_net(x, keep_prob)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate=rate)
training_operation = optimizer.minimize(loss_operation)

correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()
```

### Evaluation function

```python
def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset + BATCH_SIZE], y_data[offset:offset + BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
```

## Model Training

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)

    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: DROPOUT})

        print()
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i + 1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()

    saver.save(sess, './roadsign_net')
    print("Model saved")
```

```python
with tf.Session() as sess:
    
    # Load the weights and bias from training
    saver.restore(sess, tf.train.latest_checkpoint('.'))
    print("Model loaded")
    print()
    
    test_accuracy = evaluate(X_test, y_test)
    print("Test Accuracy = {:.3f}".format(test_accuracy))
```
